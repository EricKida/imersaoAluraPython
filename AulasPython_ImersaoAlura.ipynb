```python
df['contrato'].value_counts()
# método que calcula a frequencia de 'contrato' de cada categoria
# FT - Fulltime  CT - Contract   PT - Part-time   FL - Freelancer
```

```python
df['remoto'].value_counts()
# Quantidade de funcionarios em cada categoria de 'remoto'
# 0 - Presencial  100 - Remoto  50 - Hibrido
```

```python
contrato_traduzido = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato_traduzido)
df['contrato'].value_counts()
```

```python
tamanho_empresa_traduzido = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanhoEmpresa'] = df['tamanhoEmpresa'].replace(tamanho_empresa_traduzido)
df['tamanhoEmpresa'].value_counts()
```

```python
remoto_traduzido = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}

df['remoto'] = df['remoto'].replace(remoto_traduzido)
df['remoto'].value_counts()
```

## **Aula 2 -** Aprenda a Limpar e Preparar os Dados

```python
df.isnull()
# verifica onde possui dados nulos, marcando true. Caso esteja com algum valor, é mercado com false
```

```python
df.isnull().sum()
# verifica e soma quantos campos estão com valores nulos
```

```python
df['ano'].unique()
# mostra quais sao os valores unicos pela coluna 'ano'
```

```python
df[df.isnull().any(axis=1)]
# mostra quais são as linhas que são true no df.isnull, que possuem NaN/nulo
```

```python
import numpy as np
# importa biblioteca que possibilita trabalhar com valores nulos

df_salarios = pd.DataFrame({
    'nome': ['Luana', 'Regiane', 'Luis', 'Pedro', 'Cassio', 'Guilherme'],
    'salario': [7500, 5400, 8000, np.nan, 2000, np.nan]
})
# cria um dataframe, onde 'nome' e 'salario' sao as colunas, o que esta dentro de [] são os campos dessas colunas e np.nan é um valor nulo

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# preenche onde possui valores nulos com a media 'mean()' da coluna especificada. round2 é um arredondamento. Pode ser uma escolha errada caso o df possua valores destoantes

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
# preenche onde possui valores nulos com a mediana 'median()' da coluna especificada.

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terca', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
# na coluna preenchido_ffill, ele completa onde possui valores nulos com o valor anterior
# não pode ser usado se o valor anterior dos campos nulos tanbém forem nulos. Nesse caso, a primeira linha não poderia ser nulo

df_temperaturas
```

```python
df_temperaturas['Preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
# na coluna preenchido_ffill, ele completa onde possui valores nulos com o proximo valor
# não pode ser usado se o proximo valor dos campos nulos tanbém forem nulos. Nesse caso, o nulo não poderia ser a ultima linha

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Luana', 'Regiane', 'Luis', 'Pedro', 'Cassio', 'Guilherme'],
    'cidade': ['São Paulo', np.nan, 'Belo Horizonte', 'Porto Alegre', np.nan, 'Curitiba']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Nao informado')
# preenche com 'nao informado' onde possui valores NaN na coluna 'cidade'

df_cidades
```

```python
import pycountry

def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None
# Função para converter ISO-2 para ISO-3

df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)
# Criar nova coluna com código ISO-3

df_ds = df_limpo[df_limpo['cargo'] == 'Data Engineer']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()
# Calcular média salarial por país (ISO-3)


fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})
# Gerar o mapa interativo

fig.show()
```
